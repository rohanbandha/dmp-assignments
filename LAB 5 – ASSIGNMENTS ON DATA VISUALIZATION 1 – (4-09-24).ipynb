```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,3,2],[2,4,2])
plt.show()
```

```python
import numpy as np
ypoints= np.array([2,4,1,5])
plt.figure()
plt.plot(ypoints,'g')
plt.show()
```

```python
plt.plot(ypoints,'o')
plt.show()
```

```python
x=[10,4,7,1]
y=[2,8,3,9]
plt.plot(x, marker='^')
plt.plot(y, marker='D')
plt.show()
```

```python
plt.plot(ypoints,marker='*',ms=15,mec='r',mfc='y')
plt.show()
```

```python
plt.plot(ypoints,'*:g',ms=15)
plt.show()
```

```python
plt.title("Title")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```
